```python
import matplotlib.pyplot as plt
from matplotlib import style 
import numpy as np
```

```python
x=[1,2,3]
y=[4,5,6]
plt.plot(x,y)

 
plt.show()
```

```python


plt.plot(x,y)


#title 
plt.title('Simple graph ')

#  labels 
plt.ylabel('Rain ')
plt.xlabel('Wind')
style.use('ggplot')


#show graph 

```

```python
style.use('ggplot')

# X,y data 
x=[5,8,10]
y=[12,16,6]

# X2,y2 data 
x2=[6,9,12]
y2=[6,15,7]

#plot canvas
plt.plot(x,y,'g',label='one',linewidth=7)#g for green line, 'one' for legend label
plt.plot(x2,y2,'c',label='two',linewidth=7)#c for cyan, line width is how fat the line

plt.legend()#puts legend
#title 
plt.title('Simple graph ')

plt.xlabel('x axis')
plt.ylabel('Y axis ')
#https://matplotlib.org/stable/gallery/color/named_colors.html
plt.grid(True,color='tab:brown')# changes grid lines color
plt.grid(True,color='k')
#show graph 
plt.show
```

# Pyplot-style

```python
x = np.linspace(0, 2, 100)
x
```

```python
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.grid(True,color='k')
plt.legend()
```

```python
plt.plot(x, x,'g', label='linear',linewidth=9)#x,x,'colour,label='**',linewidth=number
plt.plot(x, x**2, label='quadratic',linewidth=9)  # etc.
plt.plot(x, x**3, label='cubic',linewidth=9)
plt.plot(x, x**4,'y', label='polynomial',linewidth=9)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.grid(True,color='k')
plt.legend()
```
